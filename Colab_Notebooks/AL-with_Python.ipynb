# Regresion Lineal

**Video sugerido para entender la regresion:** https://www.youtube.com/watch?v=k964_uNn3l0

Ecuacion de la regresion linea: *Y= b0 + b1X*

**b0:** Valor constante

**b1:** Pendiente

## Ecuacion de minimos cuadrados
**b1 = Sumatoria[(x - xmedia)*(y - ymedia)]/ Sumatoria(x - xmedia)^2**

Para encontrar el valor de b0 se evalua en el punto de las medias de x e y, esto se hace porque nuestra curva de regresion siempre pasara por esos puntos.

ymedia = b1 + b0 * xmedia, al despejar tenemos

b0 = ymedia -b1

```
import numpy as np
import matplotlib.pyplot as plt
```

```
def estimate_b0_b1(x, y):
  n = np.size(x)
  m_x, m_y = np.mean(x), np.mean(y)
  sumatoria_xy = np.sum((x - m_x) * (y - m_y))
  sumatoria_xx = np.sum((x - m_x)**2)

  b1 = sumatoria_xy / sumatoria_xx
  b0 = m_y - b1* m_x

  return b0, b1 
```

```
def plot_regression(x, y, b):
  plt.scatter(x, y, color='g', marker='o', s=30)

  y_pred = b[0] + b[1]*x
  plt.plot(x, y_pred, color='b')

  plt.xlabel('x - Independiente')
  plt.ylabel('Y - Dependiente')
  plt.show()
```

```
def main():
  x = np.array([1,2,3,4,5])
  y = np.array([2,3,5,6,5])

  b = estimate_b0_b1(x, y)
  print(f"Los valores b0 = {b[0]}, b1 = {b[1]}")

  plot_regression(x, y, b)

if __name__ == '__main__':
  main()
```
